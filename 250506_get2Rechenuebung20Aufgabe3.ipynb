```python
import numpy as np

U_q     = 230
U_Cr    = 3500
L       = 80
f_r     = 60
w_0 = f_r * 2 * np.pi

# a) Kapazitaet: C
print("# a) Kapazitaet: C\n")

C = 1 / (L * w_0**2)

print("C      = %f nF" % (C * 10**9))

# b) Guete: Q
print("\n# b) Guete: Q\n")

Q = U_Cr / U_q

print("Q      = %f" % Q)

# c) Effektivstrom im Resonanzfall: I_rEff
print("\n# c) Effektivstrom im Resonanzfall: I_rEff\n")

R      = 1 / Q * np.sqrt(w_0**2 * L**2)
I_rEff = U_q / R

print("R      = %f Ohm" % R)
print("I_rEff = %f mA" % (I_rEff * 1E3))

# d) Frequenzen, bei denen der Effektivstrom 1/10 des Stromes
# im Resonanzfall betraegt: f_1,2(I_Eff = 1/10 I_rEff)
print("\n# d) Frequenzen, bei denen der Effektivstrom 1/10 des Stromes")
print("# im Resonanzfall betraegt: f_1,2(I_Eff = 1/10 I_rEff)\n")

w_1 = w_0 * (np.sqrt(99) / 2 / Q + np.sqrt(99 / 4 / Q**2 + 1))
print("f_1    = %f Hz" % (w_1 / 2 / np.pi))

w_2 = w_0 * (np.sqrt(99) / 2 / Q - np.sqrt(99 / 4 / Q**2 + 1))
print("f_2    = %f Hz" % (w_2 / 2 / np.pi))

Z_1 = np.sqrt(R**2 + (w_1 * L - 1 / w_1 / C)**2)
print("Z      = %f Ohm" % Z_1)
Z_2 = np.sqrt(R**2 + (w_2 * L - 1 / w_2 / C)**2)
print("Z      = %f Ohm" % Z_2)

# Alternativloesung zu (d)
print("\n# Alternativloesung zu (d)\n")

w_1 = w_0**2 * (1 / 2 * (99 / Q**2 + 2) + np.sqrt(1 / 4 * (99 / Q**2 + 2)**2 - 1))
# print("w_1**2 = %f Hz**2" % w_1)
w_1 = np.sqrt(w_1)
# print("w_1    = %f Hz" % w_1)
f_1 = w_1 / 2 / np.pi
print("f_1    = %f Hz" % f_1)

w_2 = w_0**2 * (1 / 2 * (99 / Q**2 + 2) - np.sqrt(1 / 4 * (99 / Q**2 + 2)**2 - 1))
# print("w_2**2 = %f Hz**2" % w_2)
w_2 = np.sqrt(w_2)
# print("w_2    = %f Hz" % w_2)
f_2 = w_2 / 2 / np.pi
print("f_2    = %f Hz" % f_2)

Z_1 = np.sqrt(R**2 + (w_1 * L - 1 / w_1 / C)**2)
print("Z      = %f Ohm" % Z_1)
Z_2 = np.sqrt(R**2 + (w_2 * L - 1 / w_2 / C)**2)
print("Z      = %f Ohm" % Z_2)


```
